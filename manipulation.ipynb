## 7.Faisons un graphique montrant la production de banane par année pour le Sénégal. 

```python

pd_banane_an_SEN = pd.DataFrame(data1_SEN.loc[(data1_SEN['Produit']== "Bananes")&(data1_SEN['Valeur'])&(data1_SEN['AnnÃ©e']<=2016)])
pd_banane_an_SEN1= pd_banane_an_SEN[['Produit', 'Valeur', 'AnnÃ©e']]
pd_banane_an_SEN1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
Années= pd_banane_an_SEN1['AnnÃ©e']
Production= pd_banane_an_SEN1['Valeur']
couleurs = np.random.choice(["blue", "violet", "red"],30)
plt.bar(Années,Production, color = couleurs)
plt.title("Production de banane par année pour le Sénégal")
plt.xlabel("Année")
plt.ylabel("Production")
```
