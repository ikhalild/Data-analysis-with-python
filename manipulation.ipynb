# Partie B : Données sur les crimes.

>**Importation de la base de données:**

```python
data_crimes = pd.read_csv('C:/Users/pc/Documents/Formation Python for Economists Coding Cheikh Nokho/Projet final/crime.csv', encoding="latin-1")
data_crimes
```

## 1. Comptons le nombre de crimes par district.

```python
nb_cr_dis = data_crimes.loc[:,["DISTRICT", "REPORTING_AREA"]].groupby("DISTRICT").count()
nb_cr_dis
```

## 2. Faisons un histogramme du nombre de crimes par district.

```python
%matplotlib inline   
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
plt.figure(figsize = (14 , 6))
couleurs = np.random.choice(["blue", "violet", "red","green","yellow","black","orange","purple","pink",
                            "gray","brown","#00FF00"],12)
plt.hist(nb_cr_dis.T, color = couleurs, edgecolor='black')
plt.title("Nombre de crimes par district")
plt.xlabel("Districte")
plt.ylabel("Nombre de crime")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn; seaborn.set()
DISTRICT_data= data_crimes["DISTRICT"]
REPORTING_AREA_data= data_crimes["REPORTING_AREA"]
couleurs1 = np.random.choice(["blue", "violet", "red","green","yellow","black","orange","purple","pink",
                            "gray","brown","#00FF00"],30)
plt.bar(DISTRICT_data, REPORTING_AREA_data, color= couleurs1, edgecolor='black')
plt.title("Nombre de crimes par districte")
plt.xlabel("Districte")
plt.ylabel("Nombre de crime")
```
